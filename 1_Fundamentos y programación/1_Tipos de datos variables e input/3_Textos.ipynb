## Definir strings

Las cadenas se delimitan con comillas simples `'...'`, dobles `"..."` o triples `"""..."""`.

```python
simple     = 'Hola'
doble      = "Mundo"
multilinea = """Este texto
ocupa varias líneas"""
```

### Caracteres especiales

| Secuencia | Significado |
|-----------|------------|
| `\n` | Salto de línea |
| `\t` | Tabulación |
| `\"` | Comilla doble literal |
| `\'` | Comilla simple literal |
| `\\` | Barra invertida literal |

````python
texto1 = "Hola, soy 'Pepito'"        # comillas simples dentro de dobles
texto2 = 'Hola, soy "Pepito"'        # comillas dobles dentro de simples
texto3 = "Hola, soy \"Pepito\""       # escape con \
texto4 = "Línea 1\nLínea 2\nLínea 3" # saltos de línea

print(texto1)
print(texto2)
print(texto3)
print(texto4)
````

## Concatenación y repetición

````python
nombre   = "Ana"
apellido = "García"

# Concatenación con +
completo = nombre + " " + apellido
print(completo)          # → Ana García

# Repetición con *
separador = "-" * 20
print(separador)         # → --------------------

# len() — longitud de la cadena
print(len(completo))     # → 10
````

## f-strings (cadenas formateadas)

La forma más moderna y legible de insertar variables en un texto. Se escribe poniendo `f` antes de las comillas y usando `{}` para las variables.

```python
nombre = "Ana"
edad   = 30
print(f"Me llamo {nombre} y tengo {edad} años.")
```

También puedes incluir **expresiones** y **formato** dentro de los `{}`:

````python
nombre = "Pepito"
edad   = 25
precio = 19.9876

# Variable directa
print(f"Hola, {nombre}!")

# Expresión dentro de {}
print(f"En 10 años tendrás {edad + 10} años.")

# Formato numérico: 2 decimales
print(f"Precio: ${precio:.2f}")

# Alinear texto: ancho de 10 caracteres
print(f"{'Nombre':<10} {'Edad':>5}")
print(f"{nombre:<10} {edad:>5}")
````

## Métodos útiles de strings

| Método | Descripción |
|--------|-------------|
| `.upper()` | Convierte a MAYÚSCULAS |
| `.lower()` | Convierte a minúsculas |
| `.capitalize()` | Primera letra en mayúscula |
| `.title()` | Primera letra de cada palabra en mayúscula |
| `.strip()` | Elimina espacios al inicio y final |
| `.lstrip()` / `.rstrip()` | Elimina espacios solo izquierda/derecha |
| `.replace(a, b)` | Reemplaza `a` por `b` |
| `.split(sep)` | Divide en lista usando `sep` como separador |
| `.join(lista)` | Une una lista en string usando el separador |
| `.startswith(x)` | ¿Empieza con `x`? → `True`/`False` |
| `.endswith(x)` | ¿Termina con `x`? → `True`/`False` |
| `.count(x)` | Cuenta cuántas veces aparece `x` |
| `.find(x)` | Índice de la primera aparición de `x` (-1 si no existe) |

````python
texto = "  hola, mundo!  "

print(texto.strip())             # → 'hola, mundo!'
print(texto.strip().upper())     # → 'HOLA, MUNDO!'
print(texto.strip().title())     # → 'Hola, Mundo!'
print(texto.strip().replace("mundo", "Python"))  # → 'hola, Python!'
````

````python
# split y join
frase   = "manzana,banana,cereza"
frutas  = frase.split(",")        # → ['manzana', 'banana', 'cereza']
print(frutas)

unida   = " - ".join(frutas)      # → 'manzana - banana - cereza'
print(unida)
````

````python
# Búsqueda dentro de un string
texto = "Python es genial"

print(texto.startswith("Python"))  # → True
print(texto.endswith("genial"))    # → True
print(texto.count("e"))            # → 2
print(texto.find("es"))            # → 7 (índice donde empieza)
print("genial" in texto)           # → True  (operador in)
````

## Indexación y slicing

Los strings son **secuencias**: cada carácter tiene un índice.

```
 P  y  t  h  o  n
 0  1  2  3  4  5   ← índice positivo
-6 -5 -4 -3 -2 -1   ← índice negativo
```

**Slicing:** `cadena[inicio:fin:paso]` — el índice `fin` **no se incluye**.

````python
cadena = "Python"

print(cadena[0])     # → 'P'    primer carácter
print(cadena[-1])    # → 'n'    último carácter
print(cadena[0:3])   # → 'Pyt'  del índice 0 al 2
print(cadena[2:])    # → 'thon' desde el índice 2 hasta el final
print(cadena[:4])    # → 'Pyth' desde el inicio hasta el índice 3
print(cadena[::-1])  # → 'nohtyP' cadena invertida
print(cadena[::2])   # → 'Pto'  cada 2 caracteres
````
