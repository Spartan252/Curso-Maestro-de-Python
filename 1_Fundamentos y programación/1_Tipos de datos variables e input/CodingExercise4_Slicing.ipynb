# Slicing

Al realizar una consulta en un registro hemos obtenido una cadena de texto al revés. Al parecer contiene el nombre de un alumno, la nota de un exámen y la materia. Realiza lo siguiente:

1. Voltea la cadena y guárdala en una variable llamada cadena_volteada. Puedes devolver una cadena volteada usando un tercer índice negativo con slicing tal que así: cadena[::-1]

2. Extrae el nombre y apellido del alumno y almacénalos en una variable llamada nombre.

3. Extrae la nota y almacénala en una variable llamada nota.

4. Extrae la materia y almacénala en una variable llamada materia.

5. Genera exactamente la siguiente estructura formateando las anteriores variables en una cadena llamada cadena_formateada donde nombre, nota y materia hacen referencia a las variables extraídas:

    `{nombre} ha sacado un {nota} en {materia}`

Ejemplo:

`Ramón García ha sacado un 8.1 en Matemáticas`

**Notas y consejos:**

- Todas las variables del enunciado deben existir y contener el valor correcto para pasar las pruebas.

- Utiliza **slicing** para extraer las porciones de la cadena que te interesan.

- Crea tantas **variables** como necesites para hacerlo más sencillo.

```python
# Variables del ejercicio (no las modifiques)
cadena_corrupta = "airotsiH,6.7,aícraG nómaR"

# Completa el ejercicio
cadena_volteada = ""

#Pasos
#Voltear
cadena_volteada = cadena_corrupta[::-1]
print(cadena_volteada)

#Extraer componentes
componentes = cadena_volteada.split(",")
print(componentes)

#Cada componente
nombre = componentes[0]
print(nombre)

nota = componentes[1]
print(nota)

materia = componentes[2]
print(materia)

#Generación de cadena
cadena_formateada = f"{nombre} ha sacado un {nota} en {materia}"
print(cadena_formateada)
```
