## La función `input()`

`input()` **pausa el programa** y espera a que el usuario escriba algo. Siempre devuelve un `str`.

```python
nombre = input("¿Cómo te llamas? ")
print(f"Hola, {nombre}!")
```

> ⚠️ Sin importar qué escriba el usuario, `input()` **siempre devuelve texto (`str`)**.

````python
nombre = input("¿Cómo te llamas? ")
print(f"Hola, {nombre}! Bienvenido a Python.")
````

````python
# Aunque el usuario escriba un número, el resultado es str
dato = input("Escribe un número: ")
print(type(dato))    # <class 'str'>  ← siempre string
print(dato)
````

## Convertir el tipo del input

Como `input()` siempre devuelve `str`, necesitas convertirlo si quieres hacer operaciones matemáticas:

| Conversión | Uso |
|------------|-----|
| `int(input(...))` | Cuando esperas un número entero |
| `float(input(...))` | Cuando esperas un número decimal |
| `input(...).strip()` | Para limpiar espacios antes/después |

````python
# Sin convertir → error al intentar sumar
edad_str = input("¿Cuántos años tienes? ")

# Convertir a int para operar
edad = int(edad_str)
print(f"En 10 años tendrás {edad + 10} años.")
````

````python
# Forma directa: convertir en la misma línea
precio = float(input("Precio del producto: $"))
cantidad = int(input("Cantidad: "))

total = precio * cantidad
print(f"Total a pagar: ${total:.2f}")
````

## Validar entradas

Los usuarios pueden escribir cualquier cosa. Es importante **validar** que el input sea del tipo correcto antes de usarlo.

### Métodos útiles para validar strings

| Método | Descripción | Ejemplo |
|--------|------------|--------|
| `.isdigit()` | ¿Solo dígitos? | `"42".isdigit()` → `True` |
| `.isalpha()` | ¿Solo letras? | `"Ana".isalpha()` → `True` |
| `.isalnum()` | ¿Solo letras y números? | `"abc123".isalnum()` → `True` |
| `.isnumeric()` | ¿Es numérico? | `"3.14".isnumeric()` → `False` |
| `.strip()` | Elimina espacios extremos | `" hola ".strip()` → `"hola"` |

````python
# Verificar antes de convertir
dato = input("Escribe un número entero: ")

if dato.isdigit():
    numero = int(dato)
    print(f"El doble de {numero} es {numero * 2}")
else:
    print("❌ Eso no es un número entero válido.")
````

````python
# Forma robusta: usar try/except para capturar errores de conversión
dato = input("Escribe un número (puede ser decimal): ")

try:
    numero = float(dato)
    print(f"Número válido: {numero}")
except ValueError:
    print(f"❌ '{dato}' no es un número válido.")
````

````python
# Pedir el dato hasta que sea válido (bucle while)
while True:
    dato = input("Escribe un número entero positivo: ")
    if dato.isdigit() and int(dato) > 0:
        numero = int(dato)
        break
    print("❌ Entrada inválida, intenta de nuevo.")

print(f"✅ Número aceptado: {numero}")
````

## Ejemplos prácticos

````python
# Calculadora simple
print("=== Calculadora ===")
a = float(input("Primer número:  "))
b = float(input("Segundo número: "))

print(f"\n{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")

if b != 0:
    print(f"{a} / {b} = {a / b:.4f}")
else:
    print("División: no se puede dividir entre cero")
````

````python
# Construir una lista con input
n = int(input("¿Cuántos nombres quieres ingresar? "))
nombres = []

for i in range(n):
    nombre = input(f"Nombre {i + 1}: ").strip()
    nombres.append(nombre)

print(f"\nLista de nombres: {nombres}")
print(f"Total: {len(nombres)}")
````

````python
# Menú de opciones
print("=== Menú ===")
print("1. Saludar")
print("2. Calcular cuadrado")
print("3. Salir")

opcion = input("\nElige una opción (1-3): ")

if opcion == "1":
    nombre = input("¿Tu nombre? ")
    print(f"¡Hola, {nombre}!")
elif opcion == "2":
    num = float(input("¿Qué número? "))
    print(f"{num}² = {num ** 2}")
elif opcion == "3":
    print("¡Hasta luego!")
else:
    print("❌ Opción no válida.")
````
