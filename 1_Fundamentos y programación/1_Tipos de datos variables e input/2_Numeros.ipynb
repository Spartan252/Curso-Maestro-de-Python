## Enteros y flotantes

### Enteros (`int`)
Números **sin parte decimal**. Pueden ser positivos, negativos o cero. No tienen límite de tamaño en Python.

```python
edad        = 25
temperatura = -10
poblacion   = 8_000_000_000  # _ como separador de miles
```

### Flotantes (`float`)
Números **con parte decimal**. Se representan con punto `.`

```python
precio      = 19.99
pi          = 3.14159
cientifico  = 1.5e3   # notación científica = 1500.0
```

````python
entero   = 42
flotante = 3.14

print(type(entero))    # <class 'int'>
print(type(flotante))  # <class 'float'>

# Cuando sumas int + float, el resultado es float
print(entero + flotante)        # → 45.14
print(type(entero + flotante))  # <class 'float'>
````

## Operadores aritméticos

| Operador | Operación | Ejemplo | Resultado |
|----------|-----------|---------|----------|
| `+` | Suma | `3 + 2` | `5` |
| `-` | Resta | `5 - 7` | `-2` |
| `*` | Multiplicación | `12 * 6` | `72` |
| `/` | División (siempre float) | `23 / 7` | `3.285...` |
| `//` | División entera (descarta decimales) | `23 // 7` | `3` |
| `%` | Módulo (residuo) | `6 % 4` | `2` |
| `**` | Potencia | `3 ** 2` | `9` |

````python
print(f"Suma:             {3 + 2}")
print(f"Resta:            {5 - 7}")
print(f"Multiplicación:   {12 * 6}")
print(f"División:         {23 / 7}")
print(f"División entera:  {23 // 7}")
print(f"Módulo:           {6 % 4}")
print(f"Potencia:         {3 ** 2}")
````

````python
# Operadores de asignación compuesta
x = 10
x += 5   # x = 15
x -= 3   # x = 12
x *= 2   # x = 24
x /= 4   # x = 6.0
x **= 2  # x = 36.0
print(x)
````

### Orden de operaciones (PEMDAS)

Python respeta el orden estándar de las matemáticas:

1. `()` Paréntesis
2. `**` Potencia
3. `*` `/` `//` `%` Multiplicación y divisiones
4. `+` `-` Suma y resta

````python
print(2 + 3 * 4)      # → 14  (primero 3*4)
print((2 + 3) * 4)    # → 20  (paréntesis primero)
print(2 ** 3 + 1)     # → 9   (primero 2**3)
print(10 / 2 + 3)     # → 8.0
````

## Conversión de tipos

| Función | Descripción | Ejemplo |
|---------|------------|--------|
| `int(x)` | Convierte a entero (trunca) | `int(3.99)` → `3` |
| `float(x)` | Convierte a flotante | `float(5)` → `5.0` |
| `str(x)` | Convierte a texto | `str(42)` → `"42"` |
| `round(x, n)` | Redondea a `n` decimales | `round(3.14159, 2)` → `3.14` |

````python
print(int(3.99))         # → 3    (trunca, NO redondea)
print(int(-2.7))         # → -2   (trunca hacia cero)
print(float(5))          # → 5.0
print(str(42))           # → '42'
print(round(3.14159, 2)) # → 3.14
print(round(3.14159, 0)) # → 3.0
````

## Funciones numéricas útiles

| Función | Descripción | Ejemplo |
|---------|------------|--------|
| `abs(x)` | Valor absoluto | `abs(-7)` → `7` |
| `max(a, b, ...)` | Máximo | `max(3, 8, 1)` → `8` |
| `min(a, b, ...)` | Mínimo | `min(3, 8, 1)` → `1` |
| `sum(lista)` | Suma de una lista | `sum([1,2,3])` → `6` |
| `pow(x, n)` | Potencia | `pow(2, 8)` → `256` |

````python
print(abs(-7))           # → 7
print(max(3, 8, 1))      # → 8
print(min(3, 8, 1))      # → 1
print(sum([1, 2, 3, 4])) # → 10
print(pow(2, 8))         # → 256
````

````python
# El módulo math ofrece funciones matemáticas avanzadas
import math

print(math.sqrt(16))    # → 4.0   (raíz cuadrada)
print(math.pi)          # → 3.14159...
print(math.ceil(3.2))   # → 4     (redondea hacia arriba)
print(math.floor(3.9))  # → 3     (redondea hacia abajo)
print(math.log(100, 10))# → 2.0   (log base 10 de 100)
````
