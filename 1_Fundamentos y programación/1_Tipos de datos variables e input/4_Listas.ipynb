## Crear y acceder

Las listas son **colecciones ordenadas y mutables**. Pueden contener elementos de cualquier tipo, incluso mezclados.

```python
numeros = [1, 2, 3, 4, 5]
textos  = ["Ana", "Carlos", "María"]
mixta   = [42, "hola", True, 3.14]
vacia   = []
```

````python
frutas = ["manzana", "banana", "cereza", "durazno"]

print(frutas[0])    # → 'manzana'  (primer elemento)
print(frutas[-1])   # → 'durazno'  (último elemento)
print(frutas[1:3])  # → ['banana', 'cereza']  (slicing)
print(len(frutas))  # → 4  (número de elementos)

# Modificar un elemento
frutas[1] = "kiwi"
print(frutas)
````

## Métodos útiles

| Método | Descripción |
|--------|-------------|
| `.append(x)` | Agrega `x` al **final** |
| `.insert(i, x)` | Inserta `x` en la posición `i` |
| `.extend(lista2)` | Agrega todos los elementos de `lista2` al final |
| `.remove(x)` | Elimina la **primera** ocurrencia de `x` |
| `.pop(i)` | Elimina y **devuelve** el elemento en posición `i` (por defecto el último) |
| `.index(x)` | Devuelve el índice de la primera aparición de `x` |
| `.count(x)` | Cuenta cuántas veces aparece `x` |
| `.sort()` | Ordena la lista **en su lugar** |
| `.reverse()` | Invierte la lista **en su lugar** |
| `.copy()` | Devuelve una copia independiente de la lista |
| `.clear()` | Vacía la lista |

````python
numeros = [3, 1, 4, 1, 5, 9, 2, 6]

numeros.append(7)        # agrega 7 al final
numeros.insert(0, 0)     # inserta 0 al inicio
print(numeros)

numeros.remove(1)        # elimina el primer 1
print(numeros)

extraido = numeros.pop() # saca el último elemento
print(f"Extraído: {extraido}")
print(numeros)

print(numeros.count(1))  # cuántos 1 hay
print(numeros.index(9))  # índice del 9
````

````python
numeros = [3, 1, 4, 1, 5, 9, 2, 6]

ordenada = sorted(numeros)   # devuelve nueva lista ordenada (no modifica)
print("Original: ", numeros)
print("Ordenada: ", ordenada)

numeros.sort()               # ordena en su lugar
print("Sort:     ", numeros)

numeros.sort(reverse=True)   # orden descendente
print("Desc:     ", numeros)
````

## Slicing

Igual que en los strings: `lista[inicio:fin:paso]` — el índice `fin` **no se incluye**.

````python
letras = ["a", "b", "c", "d", "e", "f"]

print(letras[1:4])   # → ['b', 'c', 'd']
print(letras[:3])    # → ['a', 'b', 'c']
print(letras[3:])    # → ['d', 'e', 'f']
print(letras[::2])   # → ['a', 'c', 'e']  cada 2
print(letras[::-1])  # → ['f', 'e', 'd', 'c', 'b', 'a']  invertida
````

## Recorrer listas

La forma más común de procesar cada elemento de una lista.

````python
frutas = ["manzana", "banana", "cereza"]

# Recorrer elementos directamente
for fruta in frutas:
    print(fruta)

print()

# Con índice usando enumerate()
for i, fruta in enumerate(frutas):
    print(f"{i}: {fruta}")
````

````python
# List comprehension — forma compacta de crear listas
numeros   = [1, 2, 3, 4, 5]

cuadrados = [n ** 2 for n in numeros]
print(cuadrados)   # → [1, 4, 9, 16, 25]

pares     = [n for n in numeros if n % 2 == 0]
print(pares)       # → [2, 4]
````

## Listas de listas (matrices)

Una lista puede contener otras listas, formando estructuras de **2 o más dimensiones**.

```python
matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
```

Para acceder se usan **dos índices**: `matriz[fila][columna]`

```
       col 0  col 1  col 2
fila 0 [  1,    2,    3  ]
fila 1 [  4,    5,    6  ]
fila 2 [  7,    8,    9  ]
```

````python
matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(matriz[0][2])   # → 3   (fila 0, columna 2)
print(matriz[1])      # → [4, 5, 6]  (fila completa)
print(matriz[2][0])   # → 7   (fila 2, columna 0)

# Modificar un valor
matriz[0][0] = 99
print(matriz[0])      # → [99, 2, 3]
````

````python
# Recorrer todos los elementos con doble for
for fila in matriz:
    for elemento in fila:
        print(elemento, end=" ")
    print()  # salto de línea al terminar cada fila
````

````python
# Caso de uso: tabla de alumnos [nombre, edad, calificación]
alumnos = [
    ["Ana",    20, 9.5],
    ["Carlos", 22, 8.0],
    ["María",  21, 9.8]
]

print(alumnos[0][0])   # → 'Ana'
print(alumnos[2][2])   # → 9.8

# Agregar un nuevo alumno
alumnos.append(["Luis", 23, 7.5])

# Imprimir todos con formato
for alumno in alumnos:
    print(f"{alumno[0]:<10} edad: {alumno[1]}  cal: {alumno[2]}")
````
