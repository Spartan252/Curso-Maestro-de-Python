## Tipos de datos

| Tipo | Descripción | Ejemplo |
|------|------------|--------|
| `int` | Números enteros | `42` |
| `float` | Números decimales | `3.14` |
| `str` | Cadenas de texto | `"Hola"` |
| `bool` | Valores lógicos | `True` / `False` |
| `list` | Colecciones ordenadas y mutables | `[1, 2, 3]` |
| `tuple` | Colecciones ordenadas e inmutables | `(1, 2, 3)` |

Usa `type()` para saber el tipo de cualquier valor.

````python
print(type(42))       # <class 'int'>
print(type(3.14))     # <class 'float'>
print(type("Hola"))   # <class 'str'>
print(type(True))     # <class 'bool'>
print(type([1,2,3]))  # <class 'list'>
````

## Variables

Las variables son **espacios en memoria** que almacenan un valor. En Python no necesitas declarar el tipo.

**Reglas:** solo letras, números y `_` · no empezar con número · usar `snake_case`

```python
mes    = 12
nombre = "Pepito"
precio = 19.99
activo = True
```

````python
# Operadores de asignación compuesta
x = 10
x += 5   # x = 15
x -= 3   # x = 12
x *= 2   # x = 24
print(x)
````

## Números

```python
entero   = 42
decimal  = 3.14
resultado = 10 / 3   # → 3.333... (float)
residuo   = 10 % 3   # → 1
potencia  = 2 ** 8   # → 256
```
