## ¿Qué es una expresión anidada?

Una **expresión anidada** contiene otras expresiones dentro de sí misma. Es decir, el resultado de una expresión se usa como operando de otra.

```python
# Simple
3 + 2

# Anidada: el resultado de (3 + 2) se multiplica por 4
(3 + 2) * 4

# Más anidada
((3 + 2) * 4) - (10 / 2)
```

Python evalúa las expresiones **de adentro hacia afuera** y respeta el orden de precedencia.

````python
# Sin anidar
print(3 + 2 * 4)        # → 11  (primero * luego +)

# Anidada con paréntesis
print((3 + 2) * 4)      # → 20  (paréntesis primero)

# Doble anidación
print(((3 + 2) * 4) - (10 / 2))   # → 15.0
````

## Orden de evaluación (precedencia)

Cuando hay múltiples operadores, Python sigue este orden (de mayor a menor prioridad):

| Prioridad | Operador | Descripción |
|-----------|----------|-------------|
| 1 (mayor) | `()` | Paréntesis |
| 2 | `**` | Potencia |
| 3 | `+x`, `-x`, `not x` | Unarios |
| 4 | `*`, `/`, `//`, `%` | Multiplicación y divisiones |
| 5 | `+`, `-` | Suma y resta |
| 6 | `==`, `!=`, `>`, `<`, `>=`, `<=` | Relacionales |
| 7 | `not` | Negación lógica |
| 8 | `and` | Conjunción lógica |
| 9 (menor) | `or` | Disyunción lógica |

> Operadores del mismo nivel se evalúan de **izquierda a derecha** (excepto `**` que va de derecha a izquierda).

````python
# Potencia antes que multiplicación
print(2 ** 3 * 2)       # → 16   (primero 2**3=8, luego 8*2)
print(2 ** (3 * 2))     # → 64   (primero 3*2=6, luego 2**6)

# Multiplicación antes que suma
print(2 + 3 * 4)        # → 14
print((2 + 3) * 4)      # → 20

# Relacionales antes que lógicos
print(5 > 3 and 2 < 4)  # → True  (evalúa 5>3 y 2<4, luego and)
````

## Anidación con paréntesis

Los paréntesis son la herramienta principal para **controlar el orden** de evaluación y para **mejorar la legibilidad**.

```python
# Difícil de leer
resultado = a+b*c-d/e**2

# Mucho más claro
resultado = a + (b * c) - (d / (e ** 2))
```

````python
# Ejemplo: promedio de notas
nota_1 = 8.5
nota_2 = 9.0
nota_3 = 7.5

# Sin paréntesis → resultado incorrecto
promedio_mal = nota_1 + nota_2 + nota_3 / 3
print(f"Mal:  {promedio_mal}")   # → 20.0  (solo divide nota_3)

# Con paréntesis → correcto
promedio_bien = (nota_1 + nota_2 + nota_3) / 3
print(f"Bien: {promedio_bien}")  # → 8.33...
````

````python
# Paréntesis anidados — Python evalúa de adentro hacia afuera
resultado = ((2 + 3) * (4 - 1)) ** 2
#              ↑5       ↑3         → 5*3=15 → 15**2=225
print(resultado)  # → 225

# Fórmula cuadrática: (-b + sqrt(b²-4ac)) / 2a
import math
a, b, c = 1, -5, 6
x = (-b + math.sqrt(b**2 - 4*a*c)) / (2*a)
print(f"x = {x}")  # → 3.0
````

## Mezcla de operadores aritméticos, relacionales y lógicos

Python permite combinar los tres tipos en una sola expresión.
El flujo de evaluación es:

```
1. Aritméticos  →  2. Relacionales  →  3. Lógicos
    (número)           (bool)             (bool)
```

````python
edad = 20
saldo = 1500
bloqueado = False

# Python evalúa paso a paso:
# 1. edad >= 18          → True
# 2. saldo > 1000        → True
# 3. not bloqueado       → True
# 4. True and True       → True
# 5. True and True       → True
puede_operar = edad >= 18 and saldo > 1000 and not bloqueado
print(puede_operar)   # → True

# Otro ejemplo con or
es_fin_de_semana = True
es_festivo = False
es_vacaciones = False

dia_libre = es_fin_de_semana or es_festivo or es_vacaciones
print(dia_libre)      # → True
````

````python
# Validación de contraseña: largo >= 8 Y contiene número
password = "segura123"

largo_ok  = len(password) >= 8
tiene_num = any(c.isdigit() for c in password)

es_valida = largo_ok and tiene_num
print(f"Contraseña válida: {es_valida}")  # → True
````

## Cortocircuito lógico

Python es **perezoso**: en expresiones `and`/`or`, deja de evaluar en cuanto conoce el resultado.

| Expresión | Comportamiento |
|-----------|----------------|
| `False and ...` | No evalúa lo que sigue (ya es `False`) |
| `True or ...` | No evalúa lo que sigue (ya es `True`) |

Esto es útil para **evitar errores** y hacer código más eficiente.

````python
# Sin cortocircuito esto daría ZeroDivisionError
divisor = 0

# Gracias al cortocircuito, si divisor == 0 es True, no evalúa 10/divisor
if divisor == 0 or 10 / divisor > 2:
    print("División no segura o resultado grande")

# Con and: si la primera condición es False, la segunda no se evalúa
lista = []
if len(lista) > 0 and lista[0] == "hola":
    print("El primer elemento es hola")
else:
    print("Lista vacía o primer elemento diferente")  # → esto se imprime
````
