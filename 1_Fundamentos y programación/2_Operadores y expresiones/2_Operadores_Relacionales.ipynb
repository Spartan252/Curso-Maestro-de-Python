## Operadores relacionales

Son operadores para comparar dos valores, dependiendo del resulatdo de la comparación, puede regresar True o False.

<a id="operadores"></a>

---

| Operador | Operación | Ejemplo | Resultado |
|----------|-----------|---------|----------|
| `==` | Igual que | `3 == (1+2)` | `True` |
| `!=` | Diferente que | `5 != 7` | `True` |
| `>` | Mayor que | `3 > 1` | `True` |
| `<` | Menor que | `15 < 24` | `True` |
| `>=` | Mayor o igual que | `45 >= 23` | `True` |
| `<=` | Menor o igual que | `3 <= 3` | `True` |
| `**` | Potencia | `3 ** 2` | `9` |

```python
l1 = [1,2,3,4,5]
l2 = [6,7,8,9,10]
l3 = [2,4,5,3,1]

print(f"la lista {l1} es igual a la lista {l3}? {l1 == l3}")

l3.sort()
print(f"la lista {l1} es igual a la lista {l3}? {l1 == l3}")
```
