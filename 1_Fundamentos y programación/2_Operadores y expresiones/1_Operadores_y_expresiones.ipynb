## ¿Qué es un operador?

Un operador es un símbolo que realiza una **operación** sobre uno o más valores llamados **operandos**.

```
Operando  [operador]  Operando
   10          +         5
```

Al conjunto completo se le llama **expresión**, y siempre produce un resultado.

### Tipos de operadores en Python

| Tipo | Ejemplos | Resultado |
|------|---------|----------|
| **Aritméticos** | `+`, `-`, `*`, `/`, `**`, `%` | Número |
| **Relacionales** | `==`, `!=`, `>`, `<`, `>=`, `<=` | `True` / `False` |
| **Lógicos** | `and`, `or`, `not` | `True` / `False` |
| **Asignación** | `=`, `+=`, `-=`, `*=`, `/=` | Modifica variable |

## Tipos de expresiones

- **Aritmética** — ambos operandos son valores literales:
  ```python
  2 + 5        # → 7
  -1.4 * 54   # → -75.6
  ```

- **Algebraica** — al menos un operando es una variable:
  ```python
  radio * 3.14
  (nota_1 + nota_2) / 2
  ```

- **Relacional** — compara dos valores, siempre devuelve `bool`:
  ```python
  edad >= 18    # → True o False
  nombre == "Ana"
  ```

- **Lógica** — combina condiciones booleanas:
  ```python
  edad >= 18 and pais == "MX"
  ```

````python
# Expresión aritmética
print(2 + 5)           # → 7

# Expresión algebraica
radio = 5
print(radio * 3.14)    # → 15.7

# Expresión relacional
print(10 > 3)          # → True

# Expresión lógica
edad = 20
print(edad >= 18 and edad < 65)  # → True
````

## Operadores relacionales

