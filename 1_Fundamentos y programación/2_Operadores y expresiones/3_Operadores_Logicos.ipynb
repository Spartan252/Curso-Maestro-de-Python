## Operador Not

- Negación lógica (No)
- Sólo afecta a los tipos lógicos (True, False)

**Solo requiere un operando o una expresión**



| Ejemplo | Resultado |
|---------|----------|
| `not True` | `False` |
| `not False` | `True` |
| `not True == False` | `True` |

```python
a = 10

while not a == 0:
    print(f"{a} no es igual a 10")
    a -= 1
```

## Operador And

- Operador de **conjunción**
- Unión lógica (Y)

| Ejemplo | Resultado |
|---------|----------|
| `True and True` | `True` |
| `True and False` | `False` |
| `False and False` | `False` |

```python
b = 10
c = 23
print(f"Es {b} >= 8 Y {c} == 23? {b >= 8 and c == 23}")


edad = 23
if edad <= 13 and edad >= 0:
    print("Eres menor")
elif edad > 13 and edad < 18:
    print("Eres adolecente")
else:
    print("Eres mayor de edad")
```

## Operador Or

- Operador de **disyunción**
- Separación lógica (O)

| Ejemplo | Resultado |
|---------|----------|
| `True or True` | `True` |
| `True or False` | `True` |
| `False or False` | `False` |

```python
nombre_1 = "Miguel"
nombre_2 = "Luis"
print(f"¿Es su nombre Pedro o Luis? {nombre_1 == 'Pedro' or nombre_2 == 'Luis'}")

nombres = ["Pedro", "Alex", "Marta"]

for nombre in nombres:
    if nombre in ['Pedro', 'Luis']:
        print(f"Bienvenido a la puerta 1: {nombre}")
    elif nombre in ['Alex', 'Alberto']:
        print(f"Bienvenido a la puerta 2: {nombre}")
    elif nombre in ['Maria', 'Sofia']:
        print(f"Bienvenido a la puerta 3: {nombre}")
    else:
        print(f"{nombre} no puede entrar a ninguna puerta")
```
