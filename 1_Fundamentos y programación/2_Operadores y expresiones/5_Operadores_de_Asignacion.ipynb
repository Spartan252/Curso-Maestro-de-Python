## Asignación simple `=`

El operador `=` **guarda** un valor en una variable. No es una comparación (para comparar se usa `==`).

```python
nombre = "Ana"       # guarda el string "Ana"
edad   = 25          # guarda el entero 25
precio = 19.99       # guarda el float 19.99
activo = True        # guarda el bool True
```

> ⚠️ `=` asigna · `==` compara

````python
x = 10
print(x)         # → 10

x = 20           # reasignar: x ahora vale 20
print(x)         # → 20

x = x + 5       # usar el valor actual de x para calcular el nuevo
print(x)         # → 25

# Diferencia entre = y ==
x = 10           # asignación
print(x == 10)   # → True  (comparación)
````

## Asignación compuesta

Son atajos que **combinan una operación con la asignación**. En lugar de escribir `x = x + 5` se escribe `x += 5`.

| Operador | Equivale a | Descripción |
|----------|-----------|-------------|
| `x += n` | `x = x + n` | Suma y asigna |
| `x -= n` | `x = x - n` | Resta y asigna |
| `x *= n` | `x = x * n` | Multiplica y asigna |
| `x /= n` | `x = x / n` | Divide y asigna (resultado float) |
| `x //= n` | `x = x // n` | División entera y asigna |
| `x %= n` | `x = x % n` | Módulo y asigna |
| `x **= n` | `x = x ** n` | Potencia y asigna |

````python
x = 10
print(f"Inicio:            x = {x}")

x += 5
print(f"Después de += 5:   x = {x}")   # → 15

x -= 3
print(f"Después de -= 3:   x = {x}")   # → 12

x *= 2
print(f"Después de *= 2:   x = {x}")   # → 24

x /= 4
print(f"Después de /= 4:   x = {x}")   # → 6.0

x **= 2
print(f"Después de **= 2:  x = {x}")   # → 36.0

x %= 10
print(f"Después de %= 10:  x = {x}")   # → 6.0
````

## Asignación múltiple

Python permite asignar valores a varias variables en una sola línea.

````python
# Asignar el mismo valor a varias variables
a = b = c = 0
print(a, b, c)        # → 0 0 0

# Asignar valores distintos en una línea (unpacking)
x, y, z = 1, 2, 3
print(x, y, z)        # → 1 2 3

# Intercambiar valores sin variable temporal
a, b = 10, 20
print(f"Antes:   a={a}, b={b}")
a, b = b, a
print(f"Después: a={a}, b={b}")  # → a=20, b=10
````

````python
# Unpacking desde una lista o tupla
coordenadas = [4.5, 9.2]
lat, lon = coordenadas
print(f"Latitud: {lat}, Longitud: {lon}")

# Con * para capturar el resto
primero, *resto = [1, 2, 3, 4, 5]
print(primero)   # → 1
print(resto)     # → [2, 3, 4, 5]
````

## Casos de uso comunes

````python
# Contador
contador = 0
for _ in range(5):
    contador += 1
print(f"Conté hasta: {contador}")   # → 5
````

````python
# Acumulador de suma
numeros = [10, 25, 8, 42, 17]
total = 0

for n in numeros:
    total += n

print(f"Suma total: {total}")         # → 102
print(f"Promedio:   {total / len(numeros):.1f}")  # → 20.4
````

````python
# Carrito de compras
carrito = [
    ("Manzanas", 3, 15.50),   # (producto, cantidad, precio_unitario)
    ("Leche",    2, 22.00),
    ("Pan",      1, 35.00)
]

subtotal = 0

for producto, cantidad, precio in carrito:
    subtotal += cantidad * precio
    print(f"  {producto:<12} ${cantidad * precio:.2f}")

iva = subtotal * 0.16
total = subtotal + iva

print(f"{'Subtotal:':<14} ${subtotal:.2f}")
print(f"{'IVA (16%):':<14} ${iva:.2f}")
print(f"{'Total:':<14} ${total:.2f}")
````
