# Tipos de datos y variables

## Datos básicos

Python maneja varios tipos de datos primitivos. Los más comunes son:

| Tipo | Descripción | Ejemplo |
|------|------------|--------|
| `int` | Números enteros | `42` |
| `float` | Números decimales | `3.14` |
| `str` | Cadenas de texto | `"Hola"` |
| `bool` | Valores lógicos | `True` / `False` |
| `list` | Colecciones ordenadas y mutables | `[1, 2, 3]` |
| `tuple` | Colecciones ordenadas e inmutables | `(1, 2, 3)` |

Puedes conocer el tipo de cualquier valor usando la función `type()`.

Secciones en este notebook:
+ [Números](#numeros)
+ [Textos](#Textos)
+ [Listas](#Listas)

````python
# Ejemplos de type()
print(type(42))       # <class 'int'>
print(type(3.14))     # <class 'float'>
print(type("Hola"))   # <class 'str'>
print(type(True))     # <class 'bool'>
print(type([1,2,3]))  # <class 'list'>
````

## Variables

Las variables son **espacios en memoria** que almacenan un valor. En Python no necesitas declarar el tipo, simplemente asignas un valor con `=`.

**Reglas para nombrar variables:**
- Solo letras, números y guion bajo `_`
- No pueden empezar con un número
- Son sensibles a mayúsculas (`nombre` ≠ `Nombre`)
- Se recomienda usar `snake_case` por convención

```python
# Asignación de un valor a una variable
mes = 12
nombre = "Pepito"
lista = (12, 23.4, "Carlitos")
```

````python
var_1 = 22
var_2 = 12

resultado = var_1 + var_2
resultado
````

### Operadores de asignación compuesta

Permiten modificar el valor de una variable de forma abreviada:

| Operador | Equivale a |
|----------|------------|
| `x += 5` | `x = x + 5` |
| `x -= 5` | `x = x - 5` |
| `x *= 5` | `x = x * 5` |
| `x /= 5` | `x = x / 5` |

````python
resultado += resultado
resultado
````

## Números

### Enteros (`int`)

Son números **sin parte decimal**. Pueden ser positivos, negativos o cero. No tienen límite de tamaño en Python.

```python
edad = 25
temperatura = -10
poblacion = 8_000_000_000  # puedes usar _ como separador de miles
```

### Flotantes (`float`)

Son números **con parte decimal**. Se representan con punto `.`

```python
precio = 19.99
pi = 3.14159
cientifico = 1.5e3  # notación científica = 1500.0
```

### Operadores aritméticos

| Operador | Operación | Ejemplo | Resultado |
|----------|-----------|---------|----------|
| `+` | Suma | `3 + 2` | `5` |
| `-` | Resta | `5 - 7` | `-2` |
| `*` | Multiplicación | `12 * 6` | `72` |
| `/` | División (float) | `23 / 7` | `3.285...` |
| `//` | División entera | `23 // 7` | `3` |
| `%` | Módulo (residuo) | `6 % 4` | `2` |
| `**` | Potencia | `3 ** 2` | `9` |

````python
# Suma
print(f"Suma: {3 + 2}")

# Resta
print(f"Resta: {5 - 7}")

# Multiplicación
print(f"Multiplicación: {12 * 6}")

# División — siempre devuelve float
print(f"División: {23 / 7}")

# División entera — descarta los decimales
print(f"División entera: {23 // 7}")

# Módulo — devuelve el residuo de la división
print(f"Módulo: {6 % 4}")

# Potencia
print(f"Potencia: {3 ** 2}")
````

````python
# Conversión entre tipos numéricos
entero = int(3.99)    # → 3  (trunca, no redondea)
flotante = float(5)   # → 5.0
print(entero, flotante)
````

## Textos
<a id="Textos"></a>

### Cadenas de texto (`str`)

Las cadenas (`str`) son **secuencias de caracteres**. Se delimitan con comillas simples `'...'` o dobles `"..."`.

```python
nombre = "Ana"
saludo = 'Hola, mundo'
multilinea = """Este texto
ocupa varias líneas"""
```

### f-strings (cadenas formateadas)

Permiten **insertar variables** dentro de una cadena de forma limpia:

```python
nombre = "Ana"
edad = 30
print(f"Me llamo {nombre} y tengo {edad} años.")
# → Me llamo Ana y tengo 30 años.
```

````python
# Definir strings
nombre = "Pepito"
apellido = 'Pérez'

# Concatenación con +
nombre_completo = nombre + " " + apellido
print(nombre_completo)


Hola = "Hola soy 'Pepito'"
print(Hola)
Hola_v2 = "Hola soy \"Pepito\""
print(Hola_v2)
````

````python
# f-strings
edad = 25
print(f"Me llamo {nombre_completo} y tengo {edad} años.")
````

### Métodos útiles de strings

| Método | Descripción | Ejemplo |
|--------|------------|--------|
| `.upper()` | Convierte a mayúsculas | `"hola".upper()` → `"HOLA"` |
| `.lower()` | Convierte a minúsculas | `"HOLA".lower()` → `"hola"` |
| `.strip()` | Elimina espacios al inicio y final | `" hola ".strip()` → `"hola"` |
| `.replace(a, b)` | Reemplaza texto | `"gato".replace("g","p")` → `"pato"` |
| `.split(sep)` | Divide en lista | `"a,b,c".split(",")` → `["a","b","c"]` |
| `.len()` | *(función built-in)* Longitud | `len("hola")` → `4` |

````python
texto = "  Hola, Mundo!  "

print(texto.strip())          # Elimina espacios
print(texto.strip().upper())  # Mayúsculas
print(texto.strip().lower())  # Minúsculas
print(texto.strip().replace("Mundo", "Python"))  # Reemplazar
print(len(texto.strip()))     # Longitud
````

### Indexación y slicing

Puedes acceder a caracteres individuales o subcadenas usando corchetes `[]`:

```
 H  o  l  a
 0  1  2  3   ← índice positivo
-4 -3 -2 -1   ← índice negativo
```

````python
cadena = "Python"

print(cadena[0])    # Primer carácter → 'P'
print(cadena[-1])   # Último carácter → 'n'
print(cadena[0:3])  # Slicing → 'Pyt'
print(cadena[::2])  # Cada 2 caracteres → 'Pto'
````

## Listas
<a id="Listas"></a>

### Listas (`list`)

Las listas son **colecciones ordenadas y mutables** que pueden contener elementos de cualquier tipo.

```python
numeros = [1, 2, 3, 4, 5]
mixta   = [42, "hola", True, 3.14]
vacia   = []
```

### Métodos útiles de listas

| Método | Descripción |
|--------|-------------|
| `.append(x)` | Agrega `x` al final |
| `.insert(i, x)` | Inserta `x` en la posición `i` |
| `.remove(x)` | Elimina la primera ocurrencia de `x` |
| `.pop(i)` | Elimina y devuelve el elemento en la posición `i` |
| `.sort()` | Ordena la lista en su lugar |
| `len(lista)` | Devuelve el número de elementos |

````python
frutas = ["manzana", "banana", "cereza"]

print(frutas[0])      # Acceder → 'manzana'
frutas.append("durazno")  # Agregar
print(frutas)
frutas.remove("banana")   # Eliminar
print(frutas)
print(len(frutas))    # Longitud → 3
````
